```python


# DES Tables
IP = [58, 50, 42, 34, 26, 18, 10, 2, 60, 52, 44, 36, 28, 20, 12, 4,
      62, 54, 46, 38, 30, 22, 14, 6, 64, 56, 48, 40, 32, 24, 16, 8,
      57, 49, 41, 33, 25, 17, 9, 1, 59, 51, 43, 35, 27, 19, 11, 3,
      61, 53, 45, 37, 29, 21, 13, 5, 63, 55, 47, 39, 31, 23, 15, 7]

FP = [40, 8, 48, 16, 56, 24, 64, 32, 39, 7, 47, 15, 55, 23, 63, 31,
      38, 6, 46, 14, 54, 22, 62, 30, 37, 5, 45, 13, 53, 21, 61, 29,
      36, 4, 44, 12, 52, 20, 60, 28, 35, 3, 43, 11, 51, 19, 59, 27,
      34, 2, 42, 10, 50, 18, 58, 26, 33, 1, 41, 9, 49, 17, 57, 25]

E = [32, 1, 2, 3, 4, 5, 4, 5, 6, 7, 8, 9, 8, 9, 10, 11, 12, 13, 12, 13, 14, 15, 16, 17,
     16, 17, 18, 19, 20, 21, 20, 21, 22, 23, 24, 25, 24, 25, 26, 27, 28, 29, 28, 29, 30, 31, 32, 1]

P = [16, 7, 20, 21, 29, 12, 28, 17, 1, 15, 23, 26, 5, 18, 31, 10,
     2, 8, 24, 14, 32, 27, 3, 9, 19, 13, 30, 6, 22, 11, 4, 25]

S_BOXES = [
    [[14, 4, 13, 1, 2, 15, 11, 8, 3, 10, 6, 12, 5, 9, 0, 7],
     [0, 15, 7, 4, 14, 2, 13, 1, 10, 6, 12, 11, 9, 5, 3, 8],
     [4, 1, 14, 8, 13, 6, 2, 11, 15, 12, 9, 7, 3, 10, 5, 0],
     [15, 12, 8, 2, 4, 9, 1, 7, 5, 11, 3, 14, 10, 0, 6, 13]],

    [[15, 1, 8, 14, 6, 11, 3, 4, 9, 7, 2, 13, 12, 0, 5, 10],
     [3, 13, 4, 7, 15, 2, 8, 14, 12, 0, 1, 10, 6, 9, 11, 5],
     [0, 14, 7, 11, 10, 4, 13, 1, 5, 8, 12, 6, 9, 3, 2, 15],
     [13, 8, 10, 1, 3, 15, 4, 2, 11, 6, 7, 12, 0, 5, 14, 9]],

    [[10, 0, 9, 14, 6, 3, 15, 5, 1, 13, 12, 7, 11, 4, 2, 8],
     [13, 7, 0, 9, 3, 4, 6, 10, 2, 8, 5, 14, 12, 11, 15, 1],
     [13, 6, 4, 9, 8, 15, 3, 0, 11, 1, 2, 12, 5, 10, 14, 7],
     [1, 10, 13, 0, 6, 9, 8, 7, 4, 15, 14, 3, 11, 5, 2, 12]],

    [[7, 13, 14, 3, 0, 6, 9, 10, 1, 2, 8, 5, 11, 12, 4, 15],
     [13, 8, 11, 5, 6, 15, 0, 3, 4, 7, 2, 12, 1, 10, 14, 9],
     [10, 6, 9, 0, 12, 11, 7, 13, 15, 1, 3, 14, 5, 2, 8, 4],
     [3, 15, 0, 6, 10, 1, 13, 8, 9, 4, 5, 11, 12, 7, 2, 14]],

    [[2, 12, 4, 1, 7, 10, 11, 6, 8, 5, 3, 15, 13, 0, 14, 9],
     [14, 11, 2, 12, 4, 7, 13, 1, 5, 0, 15, 10, 3, 9, 8, 6],
     [4, 2, 1, 11, 10, 13, 7, 8, 15, 9, 12, 5, 6, 3, 0, 14],
     [11, 8, 12, 7, 1, 14, 2, 13, 6, 15, 0, 9, 10, 4, 5, 3]],

    [[12, 1, 10, 15, 9, 2, 6, 8, 0, 13, 3, 4, 14, 7, 5, 11],
     [10, 15, 4, 2, 7, 12, 9, 5, 6, 1, 13, 14, 0, 11, 3, 8],
     [9, 14, 15, 5, 2, 8, 12, 3, 7, 0, 4, 10, 1, 13, 11, 6],
     [4, 3, 2, 12, 9, 5, 15, 10, 11, 14, 1, 7, 6, 0, 8, 13]],

    [[4, 11, 2, 14, 15, 0, 8, 13, 3, 12, 9, 7, 5, 10, 6, 1],
     [13, 0, 11, 7, 4, 9, 1, 10, 14, 3, 5, 12, 2, 15, 8, 6],
     [1, 4, 11, 13, 12, 3, 7, 14, 10, 15, 6, 8, 0, 5, 9, 2],
     [6, 11, 13, 8, 1, 4, 10, 7, 9, 5, 0, 15, 14, 2, 3, 12]],

    [[13, 2, 8, 4, 6, 15, 11, 1, 10, 9, 3, 14, 5, 0, 12, 7],
     [1, 15, 13, 8, 10, 3, 7, 4, 12, 5, 6, 11, 0, 14, 9, 2],
     [7, 11, 4, 1, 9, 12, 14, 2, 0, 6, 10, 13, 15, 3, 5, 8],
     [2, 1, 14, 7, 4, 10, 8, 13, 15, 12, 9, 0, 3, 5, 6, 11]]
]

round_keys = ["010101010101010101010101010101010101010101010101" for _ in range(16)]

# Helper functions
def permute(block, table):  # performs bit permutation

    return "".join(block[i - 1] for i in table)

def xor(bin_str1, bin_str2):
    return "".join(str(int(a) ^ int(b)) for a, b in zip(bin_str1, bin_str2))

def sbox_substitution(block):
    substituted = ""
    for i in range(8): # processing the block in 8 segments
        segment = block[i*6:(i+1)*6]  # Extract 6-bit segment
        row = int(segment[0] + segment[5], 2)
        col = int(segment[1:5], 2)

        # Debugging statements
        print(f"S-Box {i+1} Input: {segment}, Row: {row}, Col: {col}")
        if row >= 4 or col >= 16:
            raise ValueError(f"Invalid S-Box indices: Row={row}, Col={col}")

        value = S_BOXES[i][row][col]
        substituted += f"{value:04b}"
    return substituted

# DES Encryption of 64-bit block
def des_encrypt_block(block, key_schedule):
    print(f"\nInitial Block: {block}")
    permuted_block = permute(block, IP)
    print(f"After Initial Permutation: {permuted_block}")
    L, R = permuted_block[:32], permuted_block[32:] # Splitting to L & R
    print(f"\nL: {L}")
    print(f"\nR: {R}")
    for i in range(16):
        expanded_R = permute(R, E) #Expansion of R
        print(f"\nRound {i+1} - Expanded R into 48 bits: {expanded_R}")

        mixed = xor(expanded_R, key_schedule[i]) #XOR with round key (key mixing)
        print(f"Round {i+1} - After Key Mixing(XOR of the expanded R with the round key): {mixed}")

        sbox_output = sbox_substitution(mixed) #S-Box substitution
        print(f"Round {i+1} - After S-Box Substitution: {sbox_output}")

        permuted_sbox_output = permute(sbox_output, P) #Permutation after S-Box
        print(f"Round {i+1} - After Permutation using P-box: {permuted_sbox_output}")

        new_R = xor(L, permuted_sbox_output) #XOR with L
        L, R = R, new_R
        print(f"Round {i+1} - After XOR of the left with the permuted sbox output: {R}")
        print(f"Round {i+1} - new L: {L}, new R: {R}")

    print(f"final_output: {R+L}")
    cipher_block = permute(R+L, FP) #Final Permutation (FP)
    print(f"\nFinal Permutation Output: {cipher_block}")
    return cipher_block



                  #-----------------------------------------------------
                            # Using DES in CBC mode

import random

# Helper methods
def generate_iv_bits():
    # Generate a 64-bit IV as a binary string
    return ''.join(str(random.randint(0, 1)) for _ in range(64))

def string_to_bits(s):
    # Convert a string to a binary string (each char is 8 bits)
    return ''.join(format(ord(char), '08b') for char in s)

def bits_to_string(bits):
    # Convert binary string back to ASCII text
    chars = [bits[i:i+8] for i in range(0, len(bits), 8)]
    return ''.join(chr(int(c, 2)) for c in chars).rstrip('\x00')  # Removes any padding null characters


                            # DES method in CBC mode
def des_encrypt_cbc(plaintext, round_keys, iv_bits):
    # Removes spaces from plaintext before dividing it into blocks
    cleaned_text = plaintext.replace(" ", "")

    # Manually splitting the plaintext into 3 blocks
    blocks = [cleaned_text[:8], cleaned_text[8:16], cleaned_text[16:]]

    # Ensure the last block is padded to 64 bits (8 bytes) if smaller than 8 characters
    if len(blocks[-1]) < 8:
        blocks[-1] += '\x00' * (8 - len(blocks[-1]))  # Pad with null characters

    previous_block = iv_bits  # IV is used for the first XOR

    print("\n--- CBC Mode Encryption Process ---")
    print(f"IV (Initial Vector): {iv_bits}\n")
    final_cipher_text = ""
    for i, block in enumerate(blocks):
        block_number = i + 1
        block_bits = string_to_bits(block)

        print(f" ~ P{block_number}: \"{block}\"")
        print(f"   Binary: {block_bits}")

        # XORing with previous ciphertext block (or IV for first block)
        xored_block = xor(block_bits, previous_block)
        print(f"   After XOR: {xored_block}")
        print("Entering the DES function:")
        # Encrypt the block using DES with 16 round keys
        Cn = des_encrypt_block(xored_block, round_keys)
        print(f"   c{i+1}: {Cn}\n")
        final_cipher_text += Cn
        # Update previous_block for next iteration
        previous_block = Cn
    final_ciphertext = bits_to_string(final_cipher_text)
    print(f"Final cipthertext is: {final_cipher_text}")
    print(f"Final cipthertext is: {final_ciphertext}")
    print("--- CBC Encryption Complete ---")

des_encrypt_cbc("Hello DES in CBC Mode", round_keys, generate_iv_bits())

#------------------------------------------------------------
```
